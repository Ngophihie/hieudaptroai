```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ file CSV
df = pd.read_csv('Social_Network.csv')
```

```python
# Kiểm tra 5 dòng đầu tiên của dữ liệu
df.head()
```

```python
# Kiểm tra thông tin về các cột trong dữ liệu
df.info()
```

```python
# Thống kê tần suất của biến phân loại 'Gender'
Gender_counts = df['Gender'].value_counts()
print(Gender_counts)
```

```python
# =>có được tần suất của từng giá trị
```

```python
# Thống kê tần suất của biến phân loại 'Ager'
Age_counts = df['Age'].value_counts()
print(Age_counts)
```

```python
# Vẽ biểu đồ cột
plt.figure(figsize=(8, 6))
sns.barplot(x=Gender_counts.index, y=Gender_counts.values)
plt.title('Tần suất các giá trị trong biến phân loại Gender')
plt.xlabel('Danh mục')
plt.ylabel('Tần suất')
plt.xticks(rotation=45)  # Xoay nhãn nếu cần
plt.show()
```

```python
# Vẽ biểu đồ tròn
plt.figure(figsize=(7, 7))
plt.pie(Gender_counts, labels=Gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Phân phối tỉ lệ của các danh mục trong biến Gender')
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.read_csv('Social_Network.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
table = pd.crosstab(df['Gender'], df['Age'])
print(table)
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ cột chồng
table.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Mối quan hệ giữa Gender và Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# Vẽ biểu đồ cột ghép
table.plot(kind='bar', figsize=(8, 6))
plt.title('Phân bố Gender theo Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```
